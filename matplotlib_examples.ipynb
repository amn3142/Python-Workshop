```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-np.pi,np.pi,256)
C,S = np.cos(X), np.sin(X)

plt.plot(X,C,color='black',linewidth=3,label='cosine')
plt.plot(X,S,color='green',linestyle='--',label='sine')

plt.xlabel('X axis',fontsize=20)
plt.ylabel('Y axis',fontsize=20)

plt.xlim([-3.0,3.0])
plt.ylim([-1.0,1.0])

plt.legend()
plt.show()
```

```python
gaussian_data = np.random.normal(size=1000000)

plt.hist(gaussian_data,bins=100)

plt.show()
```

```python
n =1000

#plt.figure(1,(10,10))

x = np.random.random_sample(n)
y = np.random.random_sample(n)

z = np.sqrt((x-np.median(x))**2.0+(y-np.median(y))**2.0)

plt.scatter(x,y,c=z)

plt.colorbar()
plt.show()
```

```python
f = np.loadtxt('./Downloads/star_dataset.txt')

x = f[:,0]
y = f[:,1]
V = f[:,2]
B = f[:,3]

plt.figure(1,(10,10))

plt.scatter(x,y,s=1,color='k')
plt.xlabel('X position')
plt.ylabel('Y position')

plt.show()

plt.figure(2,(10,10))

plt.scatter(B-V,V,s=1,color='k')
plt.ylim([26.0,11.0])
plt.xlabel('B - V',fontsize=20)
plt.ylabel('V',fontsize=20)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.show()
```
